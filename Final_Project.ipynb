## Here is the Platfroms the games were played on and how popular they were

```python
sns.catplot(x='Platform', data=df , kind='count', height = 3.5, aspect=4 )
```

## Here is the genres that show were most played

```python
sns.catplot(x='Genre', data=df , kind='count', height = 3.5, aspect=4 )
```

## These two show which company produced the most games

```python
print(df['Publisher'].value_counts())
```

```python
plt.figure(figsize=(60,8))
df['Publisher'].value_counts().plot(kind='bar', fontsize = 20)
```

```python
corr = df.corr()
fig, ax = plt.subplots(figsize=(10,10))
sns.heatmap(data=corr, square=True , annot=True, cbar=True, ax = ax)
```
